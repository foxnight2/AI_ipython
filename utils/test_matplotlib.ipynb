```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris.data.shape
```

```python
iris.target.shape
```

### plot

```python
cmaps = ['r', 'b', 'g']
for i, c in enumerate(list(set(iris.target))):
    plt.plot(iris.data[iris.target==c, 0], iris.data[iris.target==c, 1], 'o', color=cmaps[i], label=str(c))
plt.legend(loc='upper right')
```

### subplot

```python
fig = plt.figure(figsize=(10, 10))
nrow = 3
ncol = 3
for i in range(nrow):
    for j in range(ncol):
        plt.subplot(nrow, ncol, i * ncol + j + 1)
        plt.plot([0, 1], [0, 1])
plt.show()
```

### bar

```python
datas = [np.random.rand(10), np.random.rand(10)]
index = np.random.rand(10)

bins = len(datas[0])
width = 0.3
    
plt.figure(figsize=(18, 6), dpi=80)
plt.bar(x=np.arange(bins)-width/2, height=datas[0]/sum(datas[0]), width=width, label='day1')
plt.bar(x=np.arange(bins)+width/2, height=datas[1]/sum(datas[1]), width=width, label='day2')

labels = np.around(index, decimals=2)
labels = list(zip(labels, labels[1:]))
plt.xticks(ticks=range(bins), labels=labels, rotation=20)

for i in range(bins):
    plt.text(i-width, datas[0][i]/sum(datas[0])/2, np.around(datas[0][i], decimals=2), fontsize=10, rotation=20)  
    plt.text(      i, datas[1][i]/sum(datas[1])/2, np.around(datas[1][i], decimals=2), fontsize=10, rotation=20)

# plt.xlim(-1, index)
# plt.ylim(-1/4., max(height) + 0.1)
    
plt.title('name')
plt.grid(True)
plt.legend()
# plt.close()
```
