```python
# https://stock.xueqiu.com/v5/stock/f10/cn/skholderchg.json?symbol=SZ002481&extend=true&page=1&size=10
```

```python
def skholderchg(code, size=5, page=1):
    '''
    '''
    _code = ''.join(code.split('.')[::-1])
    url = 'https://stock.xueqiu.com/v5/stock/f10/cn/skholderchg.json?symbol={}&extend=true&page={}&size={}'.format(_code,
                                                                                                                   page,
                                                                                                                   size)
    response = requests.get(url, headers=headers)
    
    data = json.loads(response.text)
    
    for item in data['data']['items']:
        item['chg_date'] = time.strftime('%Y-%m-%d', time.localtime(item['chg_date']/1000))
        item['total_cash'] = item['chg_shares_num'] * item['trans_avg_price']
        
    return data
```

```python
data = skholderchg('SZ002609', 2)['data']['items']
```

```python
pools = ThreadPools(processes=64)
result = pools.imap(skholderchg, stock_basic.ts_code)
pools.close()
pools.join()
```

```python
result = list(result)
```

```python
for i, m in enumerate(result):
    
    if len([True for item in m['data']['items'] if item['chg_shares_num'] > 0]) >= 4:
        print(stock_basic.name[i])
        
```
